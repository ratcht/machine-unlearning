## Model Definition & Training

```python
import torch
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.cuda.amp import GradScaler, autocast

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "mps" if torch.mps.is_available() else "cpu")
print(f"Using device: {device}")

train_transform = transforms.Compose([
  transforms.RandomHorizontalFlip(),
  transforms.RandomRotation(10),
  transforms.RandomAffine(0, shear=10, scale=(0.8,1.2)),
  transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
  transforms.ToTensor(),
  transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Define transformations for the dataset
test_transform = transforms.Compose([
  transforms.ToTensor(),
  transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

batch_size = 128

# Load CIFAR-10 dataset
train_dataset = datasets.CIFAR10(root="../datasets/cifar10", train=True, download=True, transform=train_transform)
test_dataset = datasets.CIFAR10(root="../datasets/cifar10", train=False, download=True, transform=test_transform)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=8, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=8, pin_memory=True)
```

```python

# Define the CNN model
class CNN(nn.Module):
  def __init__(self):
    super(CNN, self).__init__()
    self.conv1 = nn.Conv2d(3, 32, 3, padding=1)  # Added padding
    self.conv2 = nn.Conv2d(32, 64, 3, padding=1)
    self.conv3 = nn.Conv2d(64, 128, 3, padding=1)  # New layer
    self.pool = nn.MaxPool2d(2)
    self.bn1 = nn.BatchNorm2d(32)
    self.bn2 = nn.BatchNorm2d(64)
    self.bn3 = nn.BatchNorm2d(128)  # New batch norm
    self.dropout1 = nn.Dropout2d(0.2)  # Reduced dropout
    self.dropout2 = nn.Dropout2d(0.3)

    # Calculate size dynamically
    dummy_input = torch.zeros(1, 3, 32, 32)
    conv_output_size = self._get_conv_output_size(dummy_input)
    
    self.fc1 = nn.Linear(conv_output_size, 256)  # Wider
    self.fc2 = nn.Linear(256, 10)

  def _get_conv_output_size(self, x):
    x = self.pool(F.relu(self.bn1(self.conv1(x))))
    x = self.pool(F.relu(self.bn2(self.conv2(x))))
    x = self.pool(F.relu(self.bn3(self.conv3(x))))
    x = self.dropout1(x)
    return torch.flatten(x, 1).shape[1]

  def forward(self, x):
    x = self.pool(F.relu(self.bn1(self.conv1(x))))
    x = self.pool(F.relu(self.bn2(self.conv2(x))))
    x = self.pool(F.relu(self.bn3(self.conv3(x))))
    x = self.dropout1(x)
    x = torch.flatten(x, 1)
    x = F.relu(self.fc1(x))
    x = self.dropout2(x)
    x = self.fc2(x)
    return F.log_softmax(x, dim=1)

# Initialize model, optimizer, and scaler
model = CNN().to(device)
optimizer = optim.Adam(model.parameters(), lr=0.001)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=3, factor=0.5)
loss_fn = nn.CrossEntropyLoss()
```

```python
# Training and testing functions
def train(model: CNN, optimizer: optim.Optimizer, loss_fn: nn.CrossEntropyLoss):
  model.train()
  total_loss = 0
  correct = 0
  for data, target in train_loader:
    data, target = data.to(device), target.to(device)
    optimizer.zero_grad()

    output: torch.Tensor = model(data)
    loss: torch.Tensor = loss_fn(output, target)

    loss.backward()
    optimizer.step()

    total_loss += loss.item()
    pred = output.argmax(dim=1, keepdim=True)
    correct += pred.eq(target.view_as(pred)).sum().item()

  accuracy = correct / len(train_loader.dataset)
  return total_loss / len(train_loader), accuracy

def test(model: CNN, loss_fn: nn.CrossEntropyLoss):
  model.eval()
  total_loss = 0
  correct = 0
  with torch.no_grad():
    for data, target in test_loader:
      data, target = data.to(device), target.to(device)
      output: torch.Tensor = model(data)
      loss: torch.Tensor = loss_fn(output, target)

      total_loss += loss.item()
      pred = output.argmax(dim=1, keepdim=True)
      correct += pred.eq(target.view_as(pred)).sum().item()

  accuracy = correct / len(test_loader.dataset)
  return total_loss / len(test_loader), accuracy
```

```python
import os
from pathlib import Path
import torch
import glob

def save_model(model: torch.nn.Module, path='models/saved_model.pt'):
  # Create directory if needed
  Path(os.path.dirname(path)).mkdir(parents=True, exist_ok=True)
  
  # If file exists, increment number until we find an unused path
  if os.path.exists(path):
    base, ext = os.path.splitext(path)
    counter = 1
    while os.path.exists(f"{base}_{counter}{ext}"):
      counter += 1
    path = f"{base}_{counter}{ext}"
    
  torch.save(model, path)
  return path

def load_model(path='models/saved_model.pt', number=None):
  if number is not None:
    base, ext = os.path.splitext(path)
    path = f"{base}_{number}{ext}"
  else:
    # Get all matching files
    base, ext = os.path.splitext(path)
    files = glob.glob(f"{base}*{ext}")
    if files:
      path = max(files)  # max will get highest numbered file
  
  if os.path.exists(path):
    model = torch.load(path)
    return model
  else:
    raise FileNotFoundError(f"No model found at {path}")
```

```python
import os
# Training loop
FORCE_TRAIN = False

if not FORCE_TRAIN and os.path.exists("models/saved_model.pt"):
  model = torch.load("models/saved_model.pt", map_location=device)
else:
  epochs = 50
  for epoch in range(1, epochs + 1):
    train_loss, train_acc = train(model, optimizer, loss_fn)
    test_loss, test_acc = test(model, loss_fn)
    print(f"Epoch {epoch}: Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, "
          f"Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}")
  save_model(model)
    
```

## Coming up with Epsilon Merging

```python
from PIL import Image
import imagehash
from torchvision.utils import save_image
import torch
import torch.nn as nn

def get_similarity(image: torch.Tensor, adv_image: torch.Tensor) -> float:
  """Hamming Distance between perceptual hashes."""
  # Save images temporarily
  save_image(image, "image_test.jpg")
  save_image(adv_image, "adv_image_test.jpg")

  # Load images
  image1 = Image.open('image_test.jpg').convert('RGB')
  image2 = Image.open('adv_image_test.jpg').convert('RGB')

  # Calculate perceptual hashes
  hash1 = imagehash.phash(image1)
  hash2 = imagehash.phash(image2)

  # Calculate the Hamming distance between the hashes
  return hash1 - hash2  # Returns an integer

def predict(input: torch.Tensor, model: nn.Module, device: torch.device, classes: list, unsqueeze: bool = False):
  if unsqueeze:
    input = input.unsqueeze(0)
  with torch.no_grad():
    input = input.to(device)
    output: torch.Tensor = model(input)
    probs = torch.softmax(output, dim=1)
    pred = output.argmax(dim=1)
    pred_class = classes[pred.item()]
    confidence = probs[0][pred].item()
    return pred_class, confidence

def merge_images(x_image: torch.Tensor, y_image: torch.Tensor, epsilon: float) -> torch.Tensor:
  adv_image = (x_image * epsilon) + (y_image * (1 - epsilon))
  adv_image = torch.clamp(adv_image, 0, 1)  # Ensure pixel values are valid
  return adv_image

def generate_adversarial_image(
  image: torch.Tensor,
  target_image: torch.Tensor,
  target_label: str,
  model: nn.Module,
  device: torch.device,
  classes: list,
  initial_epsilon: float = 0.5,
  initial_step: float = 0.05,
  num_iterations: int = 20,
  perceptual_threshold: int = 10  # Example threshold; adjust as needed
) -> torch.Tensor:
  """
  Generates an adversarial image by blending image_1 and image_2.
  
  Args:
    image_1 (torch.Tensor): The original image tensor.
    image_2 (torch.Tensor): The target adversarial image tensor.
    model (nn.Module): The pretrained model for prediction.
    device (torch.device): The device to run the model on.
    classes (list): List of class names corresponding to model outputs.
    initial_epsilon (float): Starting value for epsilon.
    initial_step (float): Step size for adjusting epsilon.
    num_iterations (int): Maximum number of iterations.
    perceptual_threshold (int): Maximum allowed Hamming distance for similarity.
  
  Returns:
    torch.Tensor: The generated adversarial image.
  """
  epsilon = initial_epsilon
  step = initial_step
  best_epsilon = None
  best_adv_image = None
  best_confidence = 0.0

  for i in range(num_iterations):
    adv_image = merge_images(image, target_image, epsilon)
    label, confidence = predict(adv_image, model, device, classes, unsqueeze=True)
    similarity = get_similarity(image, adv_image)

    print(f"Iteration {i+1}/{num_iterations}")
    print(f"  Epsilon: {epsilon:.4f}")
    print(f"  Predicted Label: {label} (Confidence: {confidence:.4f})")
    print(f"  Hamming Distance: {similarity}")

    # Check if the attack is successful and similarity is acceptable
    attack_success = label == target_label  # Assuming label_2 is target
    is_similar = similarity <= perceptual_threshold

    if attack_success and is_similar:
      print("  -> Attack successful and image is perceptually similar.")
      best_epsilon = epsilon
      best_adv_image = adv_image
      best_confidence = confidence
      # Optionally, continue to see if a higher epsilon can still maintain success
      epsilon += step
    elif attack_success and not is_similar:
      print("  -> Attack successful but image is not perceptually similar.")
      # To increase similarity, increase epsilon
      epsilon += step
    elif not attack_success and is_similar:
      print("  -> Attack failed but image is perceptually similar.")
      # To make attack stronger, decrease epsilon to make adv_image more like y_image
      epsilon -= step
    else:
      print("  -> Attack failed and image is not perceptually similar.")
      # This is a tricky case; decide based on your specific needs
      # Here, we try to decrease epsilon to strengthen the attack
      epsilon -= step

    # Clamp epsilon to [0,1]
    epsilon = max(0.0, min(1.0, epsilon))

    # Optionally, adjust step size or implement a more sophisticated search strategy

    # Early stopping if epsilon reaches boundaries
    if epsilon == 0.0 or epsilon == 1.0:
      print("  -> Epsilon reached boundary. Stopping early.")
      break

  if best_adv_image is not None:
    print(f"Best Epsilon found: {best_epsilon:.4f} with confidence {best_confidence:.4f}")
    return best_adv_image
  else:
    print("Failed to find a suitable adversarial image within the given iterations.")
    return adv_image
```

## PGD Targeted Attack

```python
def targeted_loss(output, target_class):
  """
  Computes the loss for a targeted attack by minimizing the probability of the target class.
  """
  loss_fn = nn.CrossEntropyLoss()
  target = torch.tensor([target_class]).to(output.device)
  return loss_fn(output, target)


def targeted_pgd_attack(
  model,
  image,
  target_class,
  device,
  alpha=0.01,
  epsilon=0.1,
  num_steps=40
):
  """
  Performs a targeted PGD attack on the input image to make the model predict the target class.
  
  Args:
    model (nn.Module): The pretrained model.
    image (torch.Tensor): The input image tensor.
    target_class (int): The target class index.
    device (torch.device): The device to perform computations on.
    alpha (float): Step size.
    epsilon (float): Maximum perturbation.
    num_steps (int): Number of iterations.
  
  Returns:
    torch.Tensor: The adversarial image.
  """
  # Make a copy of the image to avoid modifying the original
  adv_image = image.clone().detach().to(device)
  adv_image.requires_grad = True
  
  for step in range(num_steps):
    output = model(adv_image.unsqueeze(0))
    loss = targeted_loss(output, target_class)
    
    model.zero_grad()
    loss.backward()
    
    # Update the adversarial image by moving towards minimizing the loss
    with torch.no_grad():
      adv_image -= alpha * adv_image.grad.sign()
      # Project the perturbation to ensure it stays within the epsilon-ball
      perturbation = torch.clamp(adv_image - image, min=-epsilon, max=epsilon)
      adv_image = torch.clamp(image + perturbation, 0, 1)
    
    adv_image.requires_grad = True
  
  return adv_image
```

## Modified PGD Targeted Attack

```python
from pytorch_msssim import ssim

def combined_loss(output, target_class, adv_image, original_image, lambda_ssim=0.1):
  """
  Combines targeted loss with SSIM-based perceptual loss.
  
  Args:
    output (torch.Tensor): Model output logits.
    target_class (int): Target class index.
    adv_image (torch.Tensor): Adversarial image tensor.
    original_image (torch.Tensor): Original image tensor.
    lambda_ssim (float): Weight for the SSIM loss.
  
  Returns:
    torch.Tensor: Combined loss.
  """
  loss_fn = nn.CrossEntropyLoss()
  target = torch.tensor([target_class]).to(output.device)
  classification_loss = loss_fn(output, target)
  
  # Compute SSIM (higher SSIM means more similar)
  perceptual_loss = 1 - ssim(adv_image.unsqueeze(0), original_image.unsqueeze(0), data_range=1, size_average=True)
  
  return classification_loss + lambda_ssim * perceptual_loss


def targeted_pgd_attack_with_ssim(
  model,
  image,
  target_class,
  device,
  original_image,
  alpha=0.01,
  epsilon=0.1,
  num_steps=40,
  lambda_ssim=0.1
):
  adv_image = image.clone().detach().to(device)
  adv_image.requires_grad = True
  
  for step in range(num_steps):
    output = model(adv_image.unsqueeze(0))
    loss = combined_loss(output, target_class, adv_image, original_image, lambda_ssim)
    
    model.zero_grad()
    loss.backward()
    
    with torch.no_grad():
      adv_image -= alpha * adv_image.grad.sign()
      perturbation = torch.clamp(adv_image - original_image, min=-epsilon, max=epsilon)
      adv_image = torch.clamp(original_image + perturbation, 0, 1)
    
    adv_image.requires_grad = True
  
  return adv_image
```
